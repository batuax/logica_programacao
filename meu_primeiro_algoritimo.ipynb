**Type**

```python
x = 10
print('x', type(x))  # x é um int

nome = 'Jorge'
print('nome', type(nome))  # nome é uma string

num = "100"
num_convert = int(num)
print('num', type(num))  # num continua sendo uma string
print('num_convert', type(num_convert))  # num_convert é um int

num_float = float(num)
print('num_float', num_float)  # num_float é um float

```
