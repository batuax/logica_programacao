#Estrutura de repetição

```python
# For Range
for contador in range(10):
  print(contador)

```

```python
# For Range + Break
for contador in range(1000):
  print(contador)
  if contador == 2:
    break
```

```python
lista = ["Teste1", "Teste2"]
```

```python
# For Simples
for val in lista:
  print(val)
```

```python
# For com enumerate
for i, val in enumerate(lista):
  print(i, val)
```

* For com Operadores composto


```python
lista2 = [2, 3, 4]
```

```python
for val in lista2:
  val += 1
  print(val)
```

```python
for val in lista2:
  val *= 2
  print(val)
```

# While

```python
val = 1
while (val <6):
  val *=2
  print(val)
```

```python
# While com else
val = 16  # Se alterar o valor acima do 6 ele soma o valor total
while (val <6):
  val *=2
  print(val)
else:
  print('Valor de val é: ', val)
```
