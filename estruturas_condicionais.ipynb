#Condicional

```python
# If
if 5 > 10:
  print("verdade")
  print("5 é maior que 10")
```

```python
# If/else
num1 = 10
num2 = 5
if num1 > num2:
  print( 'O num1 é maior número')
else:
  print('O num2 é maior número')
```

```python
# if  / elif / else
num1 = 10
num2 = 10
if num1 > num2:
  print("num1 é o maior número")
elif num1 < num2:
    print("num2 é o maior número")
else:
      print("Os dois são iguais")
```

**Condicional com oiperadores lógicos**

```python
num1 = 100
num2 = 500
num3 = 2000
if (num1 > num2) and (num1 > num3):
  print('num1 é o maior número')
elif num2 < num3:
  print('o num3 é o maior número')
else:
  print('O num1 não é o maior número')
```
